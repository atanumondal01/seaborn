```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Certainly! Here are five plots that can be created using the Seaborn library, along with their uses:

1. Scatter Plot:
   Uses: It visualizes the relationship between two variables, showing how one variable changes in relation to
    another. It helps in identifying patterns, correlations, or clusters in the data.

2. Bar Plot:
   Uses: It represents categorical data with rectangular bars, making it easy to compare and display data across
    different categories. It is useful for visualizing counts, proportions, or averages for different groups.

3. Box Plot:
   Uses: It displays the distribution of a continuous variable across different categories. It provides information
    about the median, quartiles, and any potential outliers in the data, helping in identifying variations and 
    comparing distributions.

4. Heatmap:
   Uses: It visualizes a matrix of data as a color-coded grid, where each cell represents a value. It is effective in 
    showing relationships, patterns, or correlations between multiple variables, making it useful for exploratory data
    analysis.

5. Line Plot:
   Uses: It displays data points connected by lines, emphasizing the trend or change over time. It is helpful in
    visualizing time series data, tracking fluctuations, and identifying patterns or trends over a specific period.

```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
print(sns.lineplot(x=fmri.timepoint,y=fmri.signal))
```

```python
flight=sns.load_dataset("flights")
```

```python
flight
```

```python
d=flight.pivot_table(index="year",columns="month",values="passengers")
```

```python
sns.heatmap(d,annot=True,fmt="d")
```
